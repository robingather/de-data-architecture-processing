```python
import pandas as pd
import numpy as np

N_SPLITS = 5
```

```python
df = pd.read_csv("spotify_songs_full.csv", index_col=0,delimiter=",").sample(frac=1)
df_len = len(df)
split_len = int(np.floor(df_len/N_SPLITS))
```

```python
df['artists'] = df['artists'].apply(lambda x: str(x).replace(";","-"))
df['track_name'] = df['track_name'].apply(lambda x: str(x).replace(",","-"))
df['album_name'] = df['album_name'].apply(lambda x: str(x).replace(",","-"))
```

```python
df.to_csv("spotify_songs_full_preproc.csv")
```

```python
for i in range(N_SPLITS):
    slice_name = "spotify-"+str(i+1)+".csv"
    df.iloc[(i*split_len):((i+1)*split_len)].to_csv("splits/"+slice_name)
```

```python
df = pd.read_csv("spotify_artists.csv", index_col=0,delimiter=",").sample(frac=1)
df = df[df['Artist Name'] != "Artist"]
df['Lead Streams'] = df['Lead Streams'].apply(lambda x: x.replace(',','')[:-6])
df['Feats'] = df['Feats'].apply(lambda x: x.replace(',','')[:-6])
df['Tracks'] = df['Tracks'].apply(lambda x: x.replace(',',''))
df.head()
```

```python
df.to_csv("spotify_artists_preproc.csv")
```

```python
# generate artificial data
df = pd.read_csv("spotify_songs_full.csv", index_col=0,delimiter=",").sample(frac=1)
df = df.drop(['track_id','artists','album_name','key','mode','time_signature'],axis=1)
df.head()
```

```python
import random
import string

amount = 30000
genres = df['track_genre'].unique()
df_synth = pd.DataFrame()
df_synth['track_name'] = [random.choice(string.ascii_letters) for i in range(amount)]
df_synth['track_genre'] = [random.choice(genres) for i in range(amount)]
df_synth['duration'] = [30000+int(random.random()*300000) for i in range(amount)]
df_synth['popularity'] = [int(random.random()*100) for i in range(amount)]
df_synth['dance'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['energy'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['loud'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['speech'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['acoustic'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['instrument'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['live'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['positive'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['tempo'] = [np.round(70+random.random()*70) for i in range(amount)]
df_synth.head()
```

```python
df_synth.to_csv("spotify-song-stream.csv",header=False)
```
