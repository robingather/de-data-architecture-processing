```python
import pandas as pd
import numpy as np
import random
import string
```

```python
# get rid of commas in songs
df = pd.read_csv("../data/spotify_songs_full.csv", index_col=0,delimiter=",").sample(frac=1)
df['artists'] = df['artists'].apply(lambda x: str(x).replace(";","-"))
df['track_name'] = df['track_name'].apply(lambda x: str(x).replace(",","-"))
df['album_name'] = df['album_name'].apply(lambda x: str(x).replace(",","-"))
df.to_csv("../data/spotify_songs_full_preproc.csv")
df.head()
```

```python
# get rid of commas in artists
df = pd.read_csv("../data/spotify_artists.csv", index_col=0,delimiter=",").sample(frac=1)
df = df[df['Artist Name'] != "Artist"]
df['Lead Streams'] = df['Lead Streams'].apply(lambda x: x.replace(',','')[:-6])
df['Feats'] = df['Feats'].apply(lambda x: x.replace(',','')[:-6])
df['Tracks'] = df['Tracks'].apply(lambda x: x.replace(',',''))
df.to_csv("../data/spotify_artists_preproc.csv")
df.head()
```

```python
df = pd.read_csv("../data/spotify_songs_full_preproc.csv", index_col=0,delimiter=",").sample(frac=1)
df = df.drop(['track_id','artists','album_name','key','mode','time_signature'],axis=1)
df.head()
```

```python
# generate synthetic data
amount = 30000
genres = df['track_genre'].unique()
df_synth = pd.DataFrame()
df_synth['track_name'] = [random.choice(string.ascii_letters) for i in range(amount)]
df_synth['track_genre'] = [random.choice(genres) for i in range(amount)]
df_synth['duration_ms'] = [30000+int(random.random()*300000) for i in range(amount)]
df_synth['popularity'] = [int(random.random()*100) for i in range(amount)]
df_synth['danceability'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['energy'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['loudness'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['speechiness'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['acousticness'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['instrumentalness'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['liveness'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['positivity'] = [np.round(random.random(),3) for i in range(amount)]
df_synth['explicit'] = [(random.random()>0.8) for i in range(amount)]
df_synth['tempo'] = [np.round(70+random.random()*70) for i in range(amount)]
df_synth.to_csv("spotify-song-stream.csv",header=False)
df_synth.head()
```
